```python
'''
InteriorsStyle :DROP
P1:
Junk, PriceIndex9, ExpeditedListing, Zip, InsurancePremiumIndex, NormalisedPopulation, BuildYear,
    PropertyAge: Impute missing values

P2:
PriceIndex8,  PriceIndex7, PriceIndex4, PriceIndex1, PriceIndex6, PriceIndex3, 
    PriceIndex5, PriceIndex2, : Impute Missning & Convert to numeric

P3:
EnvRating, Material, Agency, AreaIncomeType, Channel, PlotType, Architecture, Region, SubModel, Facade, State, 
    RegionType, PRIMEUNIT: impute missing, create dummy, 


P4 & 5:
ListDate : date type
'''
```

```python
X_train, X_test = train_test_split(df, test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
p1 = pdPipeline([
    ('var_select', VarSelector(['Junk', 'PriceIndex9', 'ExpeditedListing', 'Zip', 'InsurancePremiumIndex', 
                                'NormalisedPopulation', 'BuildYear', 'PropertyAge'])),
    ('missing_trt', MissingValues())
])

p2 = pdPipeline([
    ('var_select', VarSelector(['PriceIndex1', 'PriceIndex2', 'PriceIndex3', 'PriceIndex4', 'PriceIndex5', 
                                'PriceIndex6', 'PriceIndex7', 'PriceIndex8'])),   
    ('convert_to_num', convert_to_numeric()),
    ('missing_trt', MissingValues())
    
])

p3 = pdPipeline([
    ('var_select', VarSelector(['EnvRating', 'Material', 'Agency', 'AreaIncomeType', 'Channel', 'PlotType','Architecture'
                                           , 'Region', 'SubModel', 'Facade', 'State', 'RegionType', 'PRIMEUNIT'])),
    ('missing_trt', MissingValues()),
    ('create_dummy', convert_to_dummy(6000))
])

p4 = pdPipeline([
    ('var_select', VarSelector(['ListDate'])),
    ('dt_cyclic', cyclic_features())
])

p5 = pdPipeline([
    ('var_select', VarSelector(['ListDate'])),
    ('dt', datetime_conversion())
])

data_pipe = FeatureUnion([
    ('missing_num', p1),
    ('obj_to_num', p2),
    ('obj_to_dum', p3),
    ('dt_to_cyclic', p4),
    ('dt_to_conv', p5)
])
```

```python
data_pipe.fit(X_train)
```

```python
len(data_pipe.get_feature_names())
```

```python
x_train = pd.DataFrame(data=data_pipe.transform(X_train),
                    columns=data_pipe.get_feature_names())
x_test=pd.DataFrame(data=data_pipe.transform(X_test),
                    columns=data_pipe.get_feature_names())
```
