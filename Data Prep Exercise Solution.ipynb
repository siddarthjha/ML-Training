```python
import pandas as pd
import numpy as np
d=pd.DataFrame({'id':np.random.choice(range(1,100),30,replace=False),
                'x':np.random.randint(1,100,30),
                'y':np.random.randint(1,100,30)})
```

```python
max_id=d['id'][(abs(d['x']-d['y'])).idxmax()]

max_id
# since the data is randomly generated , this answer will be different for different 
# runs
```

```python
x_value=int(d['x'][d['id']==max_id])
x_value
```

```python
# summing a condition gives count of True
(d['x']<x_value).sum()
```

# 3

```python
file=r"./coupon_item.csv"
```

```python
citem=pd.read_csv(file)
```

## 3.1

* I have deleted temporary dataframes created in between. 
* It will be helpful in understandin to run this code line by line and see the intermediate results

```python
coupon_features=citem['coupon_id'].value_counts().reset_index()
```

```python
coupon_features.columns=['coupon_id','coupon_occurence_count']
```

```python
coupon_features.head()
```

## 3.2

```python
temp=citem.groupby(['coupon_id'])['item_id'].nunique().reset_index()
coupon_features=pd.merge(coupon_features,temp,on='coupon_id',how='left')
del temp
coupon_features.rename(columns={'item_id':'coupon_unique_items_count'},inplace=True)
```

```python
coupon_features.head()
```

## 3.3

```python
temp=pd.crosstab(citem['coupon_id'],citem['category'])
temp.columns=[x+'_cat_count_coupon' for x in list(temp.columns)]
temp.reset_index(inplace=True)
coupon_features=pd.merge(coupon_features,temp,on='coupon_id',how='left')
del temp
```

```python
coupon_features.head()
```

## 3.4

```python
temp=citem.groupby(['coupon_id'])['category'].nunique().reset_index()
temp.columns=['coupon_id','num_unique_category_coupon']
coupon_features=pd.merge(coupon_features,temp,on='coupon_id',how='left')
del temp
coupon_features.rename(columns={'category':'coupon_unique_category_count'},inplace=True)
```

```python
coupon_features.head()
```

## 3.5

```python
temp=pd.DataFrame(citem.groupby(['coupon_id'])['brand'].value_counts())
temp.columns=['brand_count']
temp.reset_index(inplace=True)
temp=temp.groupby(['coupon_id'])['brand'].nth(0).reset_index()
temp.columns=['coupon_id','max_freq_brand']
coupon_features=pd.merge(coupon_features,temp,on='coupon_id',how='left')
del temp
```

```python
coupon_features.head()
```

## 3.6

```python
temp=pd.DataFrame(citem.groupby(['coupon_id'])['brand'].value_counts())
temp.columns=['brand_count']
temp.reset_index(inplace=True)
```

```python
temp=pd.merge(temp,coupon_features[['coupon_id','coupon_occurence_count']],
             on=['coupon_id'],how='left')
```

```python
temp['percent_freq']=temp['brand_count']/temp['coupon_occurence_count']
temp=temp[temp['percent_freq']>=0.1]
```

```python
temp=temp['coupon_id'].value_counts().reset_index()
temp.columns=['coupon_id','high_freq_brand_count']
```

```python
coupon_features=pd.merge(coupon_features,temp,how='left',on=['coupon_id'])
del temp
```

```python
coupon_features.head()
# some coupon id might not have any high freq brands 
# due to having too many unique brands 
```

## 3.7

```python
temp=pd.DataFrame(citem.groupby(['coupon_id'])['brand'].value_counts())
temp.columns=['brand_count']
temp.reset_index(inplace=True)
temp=pd.merge(temp,coupon_features[['coupon_id','coupon_occurence_count']],
             on=['coupon_id'],how='left')
```

```python
temp1=temp.groupby(['coupon_id']).nth(0).reset_index()
temp1=temp1[['coupon_id','brand_count','coupon_occurence_count']]
temp1.rename(columns={'brand_count':'b1'},inplace=True)
```

```python
temp2=temp.groupby(['coupon_id']).nth(1).reset_index()
temp2=temp2[['coupon_id','brand_count']]
temp2.rename(columns={'brand_count':'b2'},inplace=True)
```

```python
temp1=pd.merge(temp1,temp2,on=['coupon_id'],how='left')
temp1.head()
del temp2
```

```python
temp1['diff_percent_top2_brand']=(temp1['b1']-temp1['b2'])/temp1['coupon_occurence_count']

temp1=temp1[['coupon_id','diff_percent_top2_brand']]
```

```python
coupon_features=pd.merge(coupon_features,temp1,on=['coupon_id'],how='left')
del temp1
del temp
```
